```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
samples = {}
epsilon = '8'
for src in ['orig', 'thin_fgsm_' + epsilon, 'wide_fgsm_' + epsilon, 'tutorial_fgsm_' + epsilon]:
    imgs = np.load('test_%s.npy' % src)
    print src, imgs.shape, imgs.dtype, np.min(imgs), np.max(imgs)
    samples[src] = imgs[6705]
    plt.imshow(imgs[6705] / 255)
    plt.show()
```
